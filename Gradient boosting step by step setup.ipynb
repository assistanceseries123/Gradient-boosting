```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(42)
X=np.random.rand(100,1)-0.5
y=3*X[:,0]**2+0.05*np.random.randn(100)
```

```python
df=pd.DataFrame()
```

```python
df['X']=X.reshape(100)
df['y']=y
```

```python
df
```

```python
plt.scatter(df['X'],df['y'])
plt.title('x vs y')
plt.show()
```

```python
df['pred1']=df['y'].mean()
```

```python
df
```

```python
df['res1']=df['y']-df['pred1']
```

```python
df
```

```python
plt.scatter(df['X'],df['y'])
plt.plot(df['X'],df['pred1'],color='red')
```

```python
from sklearn.tree import DecisionTreeRegressor
clf=DecisionTreeRegressor(max_leaf_nodes=8)
clf.fit(df['X'].values.reshape(100,1),df['res1'].values)
```

```python
from sklearn.tree import plot_tree
plot_tree(clf)
plt.show()
```

```python
#generating x_test
x_test=np.linspace(-0.5,0.5,500)
```

```python
y_pred=0.265458+clf.predict(x_test.reshape(500,1))
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
plt.plot(x_test,y_pred,linewidth=2,color='red')
plt.scatter(df['X'],df['y'])
plt.show()
```

```python
df['pred2']=0.265458+clf.predict(df['X'].values.reshape(100,1))
```

```python
df
```

```python
df['res2']=df['y']-df['pred2']
```

```python
df
```

```python
clf2=DecisionTreeRegressor(max_leaf_nodes=8)
clf2.fit(df['X'].values.reshape(100,1),df['res2'].values)
```

```python
plot_tree(clf2)
plt.show()
```

```python
y_pred=0.265458+sum(regressor.predict(x_test.reshape(-1,1)) for regressor in [clf,clf2])
```

```python
plt.figure(figsize=(12,5))
plt.subplot(121)
plt.plot(x_test,y_pred,linewidth=2,color='red')
plt.scatter(df['X'],df['y'])
plt.title('X Vs y')
plt.show()
```

```python
def gradient_boost(X,y,number,lr,count=1,regs=[],foo=None):

  if number == 0:
    return 
  else:
    # do gradient boosting

    if count > 1:
      y = y - regs[-1].predict(X)
    else:
      foo = y

    tree_reg = DecisionTreeRegressor(max_depth=5, random_state=42)
    tree_reg.fit(X, y)

    regs.append(tree_reg)

    x1 = np.linspace(-0.5, 0.5, 500)
    y_pred = sum(lr * regressor.predict(x1.reshape(-1, 1)) for regressor in regs)

    print(number)
    plt.figure()
    plt.plot(x1, y_pred, linewidth=2)
    plt.plot(X[:, 0], foo,"r.")
    plt.show()

    gradient_boost(X,y,number-1,lr,count+1,regs,foo=foo)
```

```python
np.random.seed(42)
X=np.random.rand(100,1)-0.5
y=3*X[:,0]**2+0.05*np.random.randn(100)
gradient_boost(X,y,5,lr=1)
```
